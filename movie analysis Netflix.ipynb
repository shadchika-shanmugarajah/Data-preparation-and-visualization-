```python
# Convert the 'release_year' column to datetime
df['release_year'] = pd.to_datetime(df['release_year'], errors='coerce')

dttypes = df.dtypes

print(df['release_year'])
print(dttypes)
```

```python
# First, convert to datetime (in case itâ€™s not already)
df['release_year'] = pd.to_datetime(df['release_year'], errors='coerce')

# Now extract just the year as an integer
df['release_year'] = df['release_year'].dt.year

# Check data types
dttypes = df.dtypes

# Print results
print(df['release_year'])
print(dttypes)
df.head()
```

```python
Droping a column
```

```python
cols=['cast','description']
df.drop(cols,axis=1,inplace=True)
df.columns 
```

```python
df.head()
```

replace the rating(label converting)

```python
df['rating'] = df['rating'].replace({
    'PG-13': 'average',
    'TV-MA': 'good',
    'TV-14':'blow average',
    'TV-PG':'bad',
    'R':'popular',
    'TV-Y7':'famous',
    'TV-Y ':'not_popular'
})
df.head()
```

```python
print(df['rating'].value_counts())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.catplot(
    y='listed_in',
    data=df,
    kind='count',
    order=df['listed_in'].value_counts().index,
    color='#4287f5'
)

plt.title('Genre Column Distribution')
plt.tight_layout()
plt.show()
```

```python
top_genres = df['listed_in'].value_counts().head(10)

sns.barplot(x=top_genres.values, y=top_genres.index, color='#34a853')
plt.title('Top 10 Genres')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()
```

heat map

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='YlGnBu')
plt.title('Correlation Heatmap')
plt.show()
```

Strip Plot â€” scatter-like plot for category vs numeric

```python
sns.stripplot(data=df, x='rating', y='duration', jitter=True, palette='mako')  # 
plt.title('Duration Spread by Rating')
plt.show()
```

```python
df['release_year'] = df['release_year'].astype('Int64')  # Ensure year is numeric
yearly = df['release_year'].value_counts().sort_index()

plt.plot(yearly.index, yearly.values, marker='o', color='purple')
plt.title('Movies Released Over the Years')
plt.xlabel('Year')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
sns.boxplot(data=df, x='rating', y='duration')  
plt.title('Duration by Rating')
plt.show()
```

```python
top5 = df['listed_in'].value_counts().head(5)
colors = sns.color_palette('Set3')

plt.pie(top5, labels=top5.index, colors=colors, autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.4))
plt.title('Top 5 Genres (Donut Style)')
plt.show()
```

```python
df['release_year'].hist(bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Release Years')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.show()
```
